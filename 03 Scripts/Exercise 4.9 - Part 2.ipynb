### Step 3

```python
# Create histogram of order_hour_of_day

histogram = df_ords_prods_cust_merge['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram illustrates the distribution of orders placed at instacart across the 24-hour day cycle. Upon closer inspection, it reveals a pattern resembling a typical bell curve, indicating a relatively normal distribution. The peak of sales activity occurs during the midday hours, specifically between hour 10 and hour 15, corresponding to peak business hours. Conversely, there is a notable decrease in sales during the early morning hours (0 to 5) and late evening hours (20 to 23).

### Step 4

```python
# Create a barchart from the loyalty_flag column

bar = df_ords_prods_cust_merge['loyalty_flag'].value_counts().plot.bar(color = ['#AAC9CE', '#7CA1D9', '#E7B5D3'])
```

### Step 6

```python
# Create a sample

df_age_dependants = small[['age', 'number_of_dependants']]
```

```python
# Create line chart

line_2 = sns.lineplot(data = df_age_dependants, x = 'age', y = 'number_of_dependants')
```

There appears to be no discernible connection between age and family dynamics, as the number of dependents oscillates randomly with age. Despite increasing age, there is no evident or distinct pattern that emerges to establish a relationship between age and the number of dependents. 

### Step 7

```python
# Create sample

df_age_income = small[['age', 'income']]
```

```python
# Create a scatterplot displaying age and income

scatterplopt = sns.scatterplot(x = 'age', y = 'income', data = df_age_income)
```

A cluster appears at the lower end of the income spectrum at about the 200000 mark across all age groups. Conversely, from approximately age 40 onwards, there seems to be a trend towards higher incomes, with a concentration of data points in the upper income brackets. This observation suggests a potential correlation between age and income, implying that income tends to increase as individuals age. Notably, the density of data points for individuals over 40 is noticeably higher on the income scale compared to those under 40, indicating a possible median income increase with age. While the scatterplot visually hints at trends such as age-related income increases, rigorous statistical analysis would be necessary to validate any correlations or nuances in the age-income relationship among the customer base.
