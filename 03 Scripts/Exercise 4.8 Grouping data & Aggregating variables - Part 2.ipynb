### Step 5

```python
# Perform an aggregation of prices

df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'median', 'min', 'max']})
```

The average price of items purchased by loyal customers of Instacart are lower than that of newer customers. The average price of items purachased appears to decrease as a customer becomes more loyal to Instacart's services.  

### Step 6

```python
# Create column generating the average price of a user's order

df_ords_prods_merge['avg_order_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Check output

df_ords_prods_merge.head(15)
```

```python
# Create a spending flag associated with a user's ID

df_ords_prods_merge.loc[df_ords_prods_merge['avg_order_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_order_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Check spending flag counts

df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check output

df_ords_prods_merge[['user_id', 'spending_flag', 'avg_order_price']].head(100)
```

### Step 7

```python
# Create column generating median frequency of days_since_prior_order

df_ords_prods_merge['median_order_freq'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Check output

df_ords_prods_merge
```

```python
# Create order frequency flag associated with a user's ID

df_ords_prods_merge.loc[df_ords_prods_merge['median_order_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_order_freq'] > 10) & (df_ords_prods_merge['median_order_freq'] <= 20) , 'order_freq_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_order_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
# Check order frequency flag counts

df_ords_prods_merge['order_freq_flag'].value_counts(dropna = False)
```

```python
# Check output

df_ords_prods_merge[['user_id', 'order_freq_flag', 'median_order_freq', 'days_since_prior_order']].head(60)
```
