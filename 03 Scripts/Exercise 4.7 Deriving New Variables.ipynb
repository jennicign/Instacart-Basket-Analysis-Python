### 1. If-Statements with User-Defined Functions

```python
# Create a subset of the df

df = df_ords_prods_merge[:1000000]
```

```python
# Define function for price labels

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High-range product'
  else: return 'Not enough data'
```

```python
# Apply price range function on subset dataframe

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check values in new column

df['price_range'].value_counts(dropna = False)
```

```python
# Find maximum label price

df['prices'].max()
```

### 2. If-Statements with the loc() Function

```python
# Use the loc() function on subset

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Use loc() function on data set - loc() runs much faster than the user-defined function b/c it applies the conditional filters before searching through the dataframe, whereas the user-defined function determiens where to set filters after

df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_ords_prods_merge['price_range_loc'].value_counts(dropna = False)
```

### 3. If-Statements with For-Loops

```python
# Find frequency of orders_day_of week

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Use for-loop function to a list for assigning day-busyness when an order was placed

result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

#### 4. Creating a New Column

```python
# Create new column in df that contains result

df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
# Create loop-function for containing new criteria 'busiest days' and ' slowest days'

result_2 = []  # Initialize result_2 as an empty list

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result_2.append("Busiest days")
    elif value == 4 or value == 3:
        result_2.append("Slowest days")
    else:
        result_2.append("Regularly busy")
```

```python
# Check list output
result_2
```

```python
# Create a new column in the df

df_ords_prods_merge['busiest_days'] = result_2
```

```python
# Use for-loop function to create list that categorizes order_hour_of_day into 'most', 'fewest', and 'average orders

result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```
